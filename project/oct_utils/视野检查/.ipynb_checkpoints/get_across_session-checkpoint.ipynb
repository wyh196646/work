```python
def down_resample(image):
    return (image[:,:,::2]+image[:,:,1::2])//2
```

```python
image=down_resample(image)
```

```python
image.shape
```

```python
def get_Crosssection(temp):
    disc=np.zeros((256,256))
    x=np.arange(0,256).reshape(1,256)
    x=np.array(np.tile(x,(256,1)))
    y=np.arange(0,256).reshape(256,1)
    y=np.array(np.tile(y,(1,256)))
    disc=np.power(x-127,2)+np.power(y-127,2)
    tag=abs(disc-3600)<60
    return temp.transpose(0,2,1)[tag,:]
```

```python
 result=get_Crosssection(image)
```

```python
result.shape
```

```python
def show_img(img):
    """ 读取并展示图片

    :param path: 图片路径
    :return:
    """
    
    #print('图片的shape:', img.shape)
    plt.imshow(img)
    plt.show()


show_img(result)
```
